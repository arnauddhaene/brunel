```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x, y = np.loadtxt('Neuron0_Potentials.txt', delimiter = ':::', unpack = True)
a, b = np.loadtxt('Neuron1_Potentials.txt', delimiter = ':::', unpack = True)
c, d = np.loadtxt('Neuron2_Potentials.txt', delimiter = ':::', unpack = True)
```

```python
plt.plot(x,y, 'r')
plt.grid()
plt.title('Neuron 0 Membrane Potential')
plt.xlabel('Time in ms')
plt.ylabel('Potential in mV')
plt.show()
```

```python
plt.plot(a,b, 'b')
plt.grid()
plt.title('Neuron 1 Membrane Potential')
plt.xlabel('Time in ms')
plt.ylabel('Potential in mV')
plt.show()
```

```python
plt.plot(c,d, 'g')
plt.grid()
plt.title('Neuron 2 Membrane Potential')
plt.xlabel('Time in ms')
plt.ylabel('Potential in mV')
plt.show()
```
