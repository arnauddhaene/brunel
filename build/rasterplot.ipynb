```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x, y = np.loadtxt('rasterdataC.gdf', delimiter = '\t', unpack = True)
```

```python
fig,ax = plt.subplots(1)
ax.plot(x,y, "|", alpha=.8)
plt.xlim([560, 700])
plt.ylabel('Neuron ID')
ax.set_xticklabels([])
plt.savefig('figc.png', dpi = 400)
plt.show()


plt.hist(x, 1000, rwidth=1, alpha=.75)
plt.xlim([560, 700])
plt.ylim([0,8])
plt.xlabel('Time in ms')
plt.ylabel('Spike amount')
plt.grid(True)
plt.savefig('histC.png', dpi = 400)
plt.show()
```
